```python
# Parent class
class Resedential_House:
    # Class attribute
    Resendential_area = 'house'

    # Initializer / Instance attributes
    def __init__(self, location, rooms, bedroom, bathroom, parking, price):
        self.location = location
        self.rooms = rooms
        self.bedroom = bedroom
        self.bathroom = bathroom
        self.parking = parking
        self.price = price

    # instance method
    def Location(self):
        return "The location of this property is {}".format(self.location)

    # instance method
    def Internal(self):
        return "This property contains {} rooms , {} bedrooms and {} bathroom.".format(self.rooms, self.bedroom, self.bathroom)

    # Child class (inherits from Residential_House() class)
    def price(self):
        return "The price of this property is {}".format(self.price)


class Appartment(Resedential_House):
    def __init__(self, location , rooms, bedroom, bathroom, parking, price , floors , kitchen, lift):
        Resedential_House.__init__(self, location , rooms, bedroom, bathroom, parking, price)
        self.floors = floors
        self.kitchen = kitchen
        self.lift = lift

    def Floors(self):
        return f"""This appartment is located in {self.location} Contains {self.floors} Floors """

    def Kitchen(self):
        return f"""Kitchen is {self.Kitchen} in this apartments """

    def Lift(self):
        return f"""Lift is {self.Lift} in this apartment"""


class Farmhouse(Resedential_House):
    def __init__(self,location, rooms, bedroom, bathroom, parking, price, swimmingPool, garden, playarea):
        Resedential_House.__init__(self, location, rooms, bedroom, bathroom, parking, price)
        self.swimmingpool = swimmingPool
        self.Garden = garden
        self.playarea = playarea

    def Description(self):
        return f"""Swimming Pool is {self.swimmingpool} , Garden is {self.Garden} , Playarea is {self.playarea} in this Farmhouse and is located in {self.location} """

y = Appartment("Karachi", 4, 4, 4, "available", 300000 , "available","available","available")
print(y.Internal())
p = Farmhouse("Karachi", 4, 2, 2, "available", 390000 , "available","available","available")
print(p.Description())
```
