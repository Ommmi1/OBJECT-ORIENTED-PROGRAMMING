```python
# QUESTION # 3
class PurchaseOrder:
    def __init__(self,order,shipdate,city,street,tzip):
        self.orderNo=order
        self.shippingdate =shipdate
        self.city = city
        self.tostreet = street
        self.tozip = tzip

class OrderLineItem(PurchaseOrder):
    def __init__(self,lineno,product,quantity):
        super().__init__(order,shipdate,city,street,tzip)
        self.lineno= lineno
        self.product = product
        self.quantity = quantity
        
class Products(OrderLineItem):
    def __init__(self,productno,description,price):
        super().__init__(lineno,product,quantity)
        self.productno = productno
        self.description = description
        self.price = price

class Stock(Products):
    def __init__(self,quantity,date):
        super().__init__(self,productno,description,price)
        self.quantity = quantity
        self.date = date

class Store(Products):
    def _init_(self,location,capacity,street,city,zipcode):
        super().__init__(self,productno,description,price)
        self.location = location
        self.capacity = capacity
        self.street = street
        self.city = city
        self.zipcode = zipcode
    
class Customer():
    def __init__(self,customerno,street,city,zipcode,phone,customername):
        self.customerno =customerno
        self.street = street
        self.city = city
        self.zipcode = zipcode
        self.phone = phone
        self.customername = customername

class Person(Customer):
    def __init__(self,personid,discount):
        super().__init__(self,customerno,street,city,zipcode,phone,customername)
        self.personid=personid
        self.discount = discount
        
class Company(Customer):
    def __init__(self,Type,taxes):
        super().__init__(self,customerno,street,city,zipcode,phone,customername)
        self.Type = Type
        self.taxes = taxes
        
class CustomerAssociation(Customer):
    def __init__(self,identification,description,customerno,percentage):
        super().__init__(self,customerno,street,city,zipcode,phone,customername)    
        self.identification = identification
        self.description = description
        self.customerno = super().customerno
        self.percentage = percentage
```
