```python
class CS_Program:
    
    def __init__(self,courses):
        
        self.courses = courses
        
    def setcourse(self,course):
        
        self.course = course
        
    def getcourse(self):
        
        return self.course
    
Omer = CS_Program('  THE TOTAL COURSES ARE \n  CS,SE,PE,EE')
Omer.setcourse('  YOUR COURSES ARE \n  CS,SE')
print(Omer.courses)
print(Omer.getcourse())
```
